$MaxDepth(root) = max(MaxDepth(root.left), MaxDepth(root.right)) + 1$

```python
import algviz

class Solution:
    def __init__(self, root) -> None:
        self.viz = algviz.Visualizer(layout=True)
        self.tree = self.viz.createGraph(root)
        self.viz.display()
        self.maxDepth(root)
        self.viz.display(1)
        self.viz.layout()

    def maxDepth(self, root):
        if root is None:
            return 0
        self.tree.markNode(algviz.cGreen, root, True)
        self.tree.markNode(algviz.cLime, root)
        self.viz.display(1)
        left_max_depth = self.maxDepth(root.left)
        right_max_depth = self.maxDepth(root.right)
        if left_max_depth > right_max_depth:
            self.tree.markNode(algviz.cOrange, root.left)
        else:
            self.tree.markNode(algviz.cOrange, root.right)
        tree_max_depth = max(left_max_depth, right_max_depth) + 1
        root.val = '{}\ndep:{}'.format(root.val, tree_max_depth)
        self.tree.markNode(algviz.cDarkBlue, root, True)
        self.viz.display()
        return tree_max_depth
```

```python
tree_nodes = algviz.parseBinaryTree([3,9,20,None,None,15,7])
Solution(tree_nodes)
```
