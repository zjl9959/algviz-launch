```python
import algviz

# Code.S
class Solution:
    # Hide.S
    def __init__(self, root):
        self.viz = algviz.Visualizer()
        self.tree = self.viz.createGraph(root, '完整二叉树')
    # Hide.E

    def evaluateTree(self, root):
        self.tree.markNode(algviz.cSpringGreen, root)   # Hide
        self.viz.display()  # Hide
        res = False
        if root.left:
            left_res = self.evaluateTree(root.left)
            right_res = self.evaluateTree(root.right)
            if root.val == 2:
                res = left_res or right_res
            elif root.val == 3:
                res = left_res and right_res
        elif root.val == 1:
            res = True
        # Hide.S
        root.val = str(res)
        self.viz.display()
        self.tree.markNode(algviz.cSilver, root, hold=True)
        self.viz.display()
        # Hide.E
        return res
# Code.E
```

```python
# Case1 from example:
tree_info = [2, 1, 3, None, None, 0, 1]
root = algviz.parseBinaryTree(tree_info)
Solution(root).evaluateTree(root)
```

```python
# Case2 from example:
tree_info = [2, 2, 3, 1, 1, 0, 1]
root = algviz.parseBinaryTree(tree_info)
Solution(root).evaluateTree(root)
```
