```python
import algviz

# Code.S
class Solution:
    # Hide.S
    def __init__(self, root):
        self.viz = algviz.Visualizer()
        self.tree = self.viz.createGraph(root, 'å®Œæ•´äºŒå‰æ ‘')
    # Hide.E

    def evaluateTree(self, root):
        self.tree.markNode(algviz.cSpringGreen, root)   # Hide
        self.viz.display()  # Hide
        res = False
        if root.left:
            left_res = self.evaluateTree(root.left)
            right_res = self.evaluateTree(root.right)
            if root.val == 2:
                res = left_res or right_res
            elif root.val == 3:
                res = left_res and right_res
        elif root.val == 1:
            res = True
        # Hide.S
        root.val = str(res)
        self.viz.display()
        self.tree.markNode(algviz.cSilver, root, hold=True)
        self.viz.display()
        # Hide.E
        return res
# Code.E
```

```python
class Solution:
    def evaluateTree(self, root):
        res = False
        if root.left:
            left_res = self.evaluateTree(root.left)
            right_res = self.evaluateTree(root.right)
            if root.val == 2:
                res = left_res or right_res
            elif root.val == 3:
                res = left_res and right_res
        elif root.val == 1:
            res = True
        return res
```
