```python
import algviz   # algviz æ˜¯ä¸€ä¸ªå¼€æºçš„ç®—æ³•åŠ¨ç”»å¼•æ“Žï¼Œä»“åº“åœ°å€è§æˆ‘ä¸»é¡µä¸­çš„ç®€ä»‹

class Allocator(object):

    def __init__(self, n):
        """
        :type n: int
        """
        self.viz = algviz.Visualizer(1, 3)
        self.mem = self.viz.createVector([None for _ in range(n)],
                                        name="æ¨¡æ‹Ÿå†…å­˜", cell_size=(30, 30))
        self.free_mem = {0: n}
        self.used_mem = dict()
        self.log = self.viz.createLogger(1, 'å½“å‰æ“ä½œ')

    def allocate(self, size, mID):
        """
        :type size: int
        :type mID: int
        :rtype: int
        """
        res = -1
        for index in sorted(self.free_mem.keys()):
            len = self.free_mem[index]
            if len >= size:
                for i in range(index, index+size):
                    self.mem[i] = mID
                    self.mem.mark(algviz.cTomato, i)
                if len - size > 0:
                    self.free_mem[index+size] = len - size
                self.free_mem.pop(index)
                if mID in self.used_mem:
                    self.used_mem[mID].append(index)
                else:
                    self.used_mem[mID] = [index]
                res = index
                break
        self.log.write('allocate({}, {}), return={}'.format(size, mID, res))
        self.viz.display()
        return res

    def free(self, mID):
        """
        :type mID: int
        :rtype: int
        """
        if mID not in self.used_mem:
            return 0
        index_list = self.used_mem.pop(mID)
        res = 0
        for index in index_list:
            left = index - 1
            while left >=0 and self.mem[left] == None:
                left -= 1
            right = index
            while right < len(self.mem) and self.mem[right] == mID:
                self.mem[right] = None
                right += 1
            res += right - index
            while right < len(self.mem) and self.mem[right] == None:
                right += 1
            self.free_mem[left+1] = right - left - 1
            k_pop_list = list()
            for k in self.free_mem.keys():
                if k > left+1 and k < right-1:
                    k_pop_list.append(k)
            for k in k_pop_list:
                self.free_mem.pop(k)
        self.log.write('free({}), return={}'.format(mID, res))
        self.viz.display()
        return res
```

```python
class Allocator(object):

    def __init__(self, n):
        """
        :type n: int
        """
        self.mem = [None for _ in range(n)]
        self.free_mem = {0: n}
        self.used_mem = dict()

    def allocate(self, size, mID):
        """
        :type size: int
        :type mID: int
        :rtype: int
        """
        res = -1
        for index in sorted(self.free_mem.keys()):
            len = self.free_mem[index]
            if len >= size:
                for i in range(index, index+size):
                    self.mem[i] = mID
                if len - size > 0:
                    self.free_mem[index+size] = len - size
                self.free_mem.pop(index)
                if mID in self.used_mem:
                    self.used_mem[mID].append(index)
                else:
                    self.used_mem[mID] = [index]
                res = index
                break
        return res

    def free(self, mID):
        """
        :type mID: int
        :rtype: int
        """
        if mID not in self.used_mem:
            return 0
        index_list = self.used_mem.pop(mID)
        res = 0
        for index in index_list:
            left = index - 1
            while left >=0 and self.mem[left] == None:
                left -= 1
            right = index
            while right < len(self.mem) and self.mem[right] == mID:
                self.mem[right] = None
                right += 1
            res += right - index
            while right < len(self.mem) and self.mem[right] == None:
                right += 1
            self.free_mem[left+1] = right - left - 1
            k_pop_list = list()
            for k in self.free_mem.keys():
                if k > left+1 and k < right-1:
                    k_pop_list.append(k)
            for k in k_pop_list:
                self.free_mem.pop(k)
        return res
```
