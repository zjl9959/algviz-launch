è¯·ä¾æ¬¡è¿è¡Œä¸‹æ–¹å•å…ƒæ ¼ä¸­çš„ä»£ç æ¥ç”Ÿæˆå¯è§†åŒ–åŠ¨ç”»ï¼ğŸ‘‡

*å¦‚æœæ‚¨åœ¨è¿è¡Œä¸­é‡åˆ°äº†æŠ¥é”™ï¼Œéº»çƒ¦æŠŠæŠ¥é”™æ—¥å¿—æäº¤åˆ°è¿™é‡Œï¼ˆ[Algviz Github issues](https://github.com/zjl9959/algviz/issues)ï¼‰ï¼Œä¹Ÿå¯ä»¥ç›´æ¥å‘æˆ‘é‚®ç®±ï¼ˆzjl9959@gmail.comï¼‰ï¼Œè°¢è°¢ï¼*

```python
import algviz

class Solution(object):
    def minOperations(self, nums):
        viz = algviz.Visualizer(1, 2)
        nums = viz.createVector(nums)
        log = viz.createLogger(1, show_line_num=False)
        viz.display()
        add_count = 0
        i = viz.createCursor(1, 'i')
        while i < len(nums):
            if nums[i] <= nums[i-1]:
                log.write('æ‰€éœ€æ“ä½œæ¬¡æ•°ä¹‹å’Œï¼š{} + ( {} - {} ) + 1 = {}'.format(
                    add_count,  nums[i-1], nums[i], add_count+nums[i-1]-nums[i]+1))
                add_count += nums[i-1]-nums[i]+1
                nums[i] = nums[i-1] + 1
            viz.display()
            i += 1
        return add_count
```

```python
nums = [1,5,2,4,1]
print(Solution().minOperations(nums))
```

```python
class Solution(object):
    def minOperations(self, nums):
        add_count = 0
        i = 1
        while i < len(nums):
            if nums[i] <= nums[i-1]:
                add_count += nums[i-1]-nums[i]+1
                nums[i] = nums[i-1] + 1
            i += 1
        return add_count

nums = [1,5,2,4,1]
print(Solution().minOperations(nums))
```
