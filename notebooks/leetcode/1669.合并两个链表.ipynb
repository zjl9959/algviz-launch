## 解题思路

## 算法演示代码

请依次运行下方单元格中的代码来生成可视化动画！👇

```python
import algviz

# Code.S
class Solution:
    def mergeInBetween(self, list1, a, b, list2):
        viz = algviz.Visualizer(1)
        li1 = viz.createGraph(list1)
        li2 = viz.createGraph(list2)
        rm0, rm1 = list1, list1
        while a > 1:
            li1.markNode(algviz.cOrange, rm0)
            viz.display()
            rm0 = rm0.next
            a-=1
        li1.markNode(algviz.cOrange, rm0, True); viz.display()
        while b > -1:
            li1.markNode(algviz.cLightBlue, rm1)
            viz.display()
            rm1 = rm1.next
            b-=1
        li1.markNode(algviz.cLightBlue, rm1, True); viz.display()
        tail = list2
        while tail.next:
            li2.markNode(algviz.cSpringGreen, tail)
            viz.display()
            tail = tail.next
        li2.markNode(algviz.cSpringGreen, tail, True); viz.display()
        rm0.next = list2
        viz.display(3)
        tail.next = rm1
        viz.display(3)
        return list1
# Code.E
```

```python
# case1 from example:
list1 = algviz.parseForwardLinkedList([0,1,2,3,4,5])
list2 = algviz.parseForwardLinkedList([1000000,1000001,1000002])
Solution().mergeInBetween(list1, 3, 4, list2)
```

## 可提交版本代码

```python
class Solution:
    def mergeInBetween(self, list1, a, b, list2):
        rm0, rm1 = list1, list1
        while a > 1:
            rm0 = rm0.next
            a-=1
        while b > -1:
            rm1 = rm1.next
            b-=1
        rm0.next = list2
        tail = list2
        while tail.next:
            tail = tail.next
        tail.next = rm1
        return list1
```
